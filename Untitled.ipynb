Описание:
Система Лотки-Вольтерры является первоначальной и простейшей системой
(усложненные системы будут рассмотрены ниже) для описания модели
«хищник-жертва», то есть популяции хищников и популяции жертв,
взаимодействующих в какой-то среде: жертвы едят растительность, хищники
— жертв:

\begin{cases}
\dot{x} & =(\alpha - \beta y)x \\
\dot{y} & =(-\gamma + \delta x)y \\
\end{cases}

,где 

* x — численность жертв (травоядных);
* y — численность хищников;
* α — вероятность того, что травоядные размножатся;
* β — вероятность того, что травоядное будет съедено хищником;
* γ — вероятность того, что хищник умрет от голода;
* δ — вероятность того, что хищнику хватит еды на дальнейшее размножение. 


Из системы сразу следует, что если жертв нет (x = 0), то хищники будут
вымирать экспоненциально с неким начальным коэффициентом (γ согласно
уравнению). 

\begin{align}
\dot{y} & = -\gamma y => y =C_1e^{-\gamma t},C_1 \in \mathbb{R} \\
\end{align}


Схожую ситуацию получаем при полном отсутствии хищников (y = 0):
   
\begin{align}
\dot{x} & = \alpha x => x =C_2e^{-\alpha t},C_2 \in \mathbb{R} \\
\end{align}

Рост жертв получается экспоненциальным с некой заранее заданной
константой (α). Стоит отметить, что в данной модели принимаются
несколько допущений:
* Количество пищи для травоядных не ограничено;
* Ни жертвы, ни хищники не эмигрируют из среды;
* Никакие другие животные не мигрируют в среду;
* Данная модель не учитывает вымирание животных по причине старения и прочих внешних воздействий. 


Найдем особые точки, которыми обладает система
\begin{align}
\begin{cases}
(\alpha - \beta y)x = 0) \\
(-\gamma + \delta x) y = 0 \\
\end{cases}
\xrightarrow[]{}
\begin{cases}
\alpha x = \beta xy \\
\gamma y = \delta xy\\
\end{cases}
\xrightarrow[]{\text{$x \neq 0$, $y \neq 0$}}
\begin{cases}
y(0) = \dfrac{\alpha}{\beta} \\
x(0) = \dfrac{\gamma}{\delta} \\
\end{cases}
\end{align}

Понятно, что при x(0) = 0, y(0) = 0 особой точкой будет как раз (0, 0), но этот случай не интересен, так как в нулевой момент времени животные обоих видов отсутствуют и, что логично, дальше не появляются. 

Гораздо более интересные вещи происходят в ненулевом случае. В зависимости от начальных параметров будет меняться особая точка — такое значение размеров популяции животных, когда обе популяции остаются неизменными и сбалансированными. 

Если же начальное условие не попадает в особую точку, фазовые кривые будут идти вокруг нее, образуя бесконечное циклическое колебание, о котором как раз и говорили Лотка и Вольтерра. То есть количество особей одного вида будет расти, другого — падать, затем наоборот, и так в течение неограниченного количества времени (в разумных пределах, конечно). 

### Напишем программу для расчета модели

```python
import matplotlib.pyplot as plt
import numpy as np
```

Создадим переменные для всех параметров.

```python
alpha = 0.42
beta = 0.14
gamma = 0.79
delta = 0.03
h = 0.01
tpoints = np.arange(0, 500, h)
x0 = 10.0
y0 = 5.0
```

Функция для расчета количества жертв.

```python
def dxdt(x,y):
    return x*(alpha - beta*y)
```

Функция для расчета количества хищников.

```python
def dydt(x,y):
    return  y*(-gamma + delta*x)
```

Функции реализующие метод Рунге - Кутты.

```python
def rk4x(x, y, h):      
    k1 = dxdt(x,y)
    k2 = dxdt(x+0.5*h,y+0.5*h*k1)
    k3 = dxdt(x+0.5*h,y+0.5*h*k2)
    k4 = dxdt(x+h,y+h*k3)
    return ((k1 + 2*k2 + 2*k3 + k4)*h)/6
```

```python
def rk4y(x, y, h):      
    k1 = dydt(x,y)
    k2 = dydt(x+0.5*h,y+0.5*h*k1)
    k3 = dydt(x+0.5*h,y+0.5*h*k2)
    k4 = dydt(x+h,y+h*k3)
    return ((k1 + 2*k2 + 2*k3 + k4)*h)/6
```

Реализация симуляции модели.

```python
xpoints, ypoints  = [], []
for t in tpoints:
    xpoints.append(x0) 
    ypoints.append(y0)
    tx,ty = x0, y0
    x0 += rk4x(tx,ty,h)
    y0 += rk4y(tx,ty,h) 
```

Численные значения точек с 50 по 60 для жертв.

```python
print(np.column_stack((tpoints[50:60],xpoints[50:60])))
```

Численные значения точек с 50 по 60 для хищников.

```python
print(np.column_stack((tpoints[50:60],ypoints[50:60])))
```

Отображение графика зависимотси популяция друг от друга.

```python
plt.plot(tpoints, xpoints, label="Жертвы")
plt.plot(tpoints, ypoints, label="Хищники")
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',
           ncol=2, mode="expand", borderaxespad=0.)
plt.style.use('seaborn')
plt.xlabel("Time")
plt.ylabel("Population")
plt.title("Lotka-Volterra Model")
plt.grid(True)
plt.show()
```
